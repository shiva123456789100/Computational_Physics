```python
import numpy as np

def is_prime(n):
    c=0
    if n==2:
        return "Yes"
    else:
        for i in range(2,int(np.sqrt(n))+1):
            if n%i==0.00:
                c=1
                return "No"
            
            else:
                c=0
    if c==0:
        return "Yes"
                

    
        
        
```

```python
is_prime(35)
```

```python
def prime(n,m):
    a=[]
    for i in range(min(n,m),max(n,m)):
        if is_prime(i)=="Yes":
            a.append(i)
    return a
```

```python
prime(9,100)
```

```python
def f(x,n):
    s=1   #As for n=0 , it should give output 0 
    l=1
    for i in range(1,n+1):
        l=l*x/i
        s=s+l
    return s
```

```python
f(1,100)
```

```python
def lebn(n):
    t=0
    s=0
    for i in range(1,int(n)+1):
        t=1/(2*i-1)*(-1)**(i+1)
        s=s+t
    return s*4

        
```

```python
def f(n):
    t=0
    s=1
    for i in range(1,int(n)+1):
        t=(2*n+1-2*i)**2/(2+t)
    return 4/(t+1)
    
```

```python
z=np.linspace(1,10,10)
data={}
for i in z:
    data[i]=np.round([lebn(i),f(i)],8)
df = pd.DataFrame(data, index =["Lebnitz's formula","Continued fraction"])


print(df)
    
    
    
```
