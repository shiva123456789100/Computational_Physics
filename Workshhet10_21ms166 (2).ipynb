```python
def area(n):#Defining the area function to compute no of points satisying the required criteria of being in the shaded region
    meet = 0
    for i in range(n):
        ta=30*np.random.rand() #ta, tb are respectively time when a and respectively turns up.
        tb=30*np.random.rand()
        if ta<tb:
            if tb-ta<5:
                meet+=1
        else:
            if ta-tb<10:
                meet+=1
    return meet/n

```

```python
lens=np.arange(20,200000,1000)#Checking a suitable value of n for which the probability is converging.
area_list=np.array([area(i) for i in lens])
plt.plot(lens,area_list)
plt.xlabel("No of trials")
plt.ylabel("Probability")
```

```python
def prob(n,d,l):
    count=0
    for i in range(n):
        com_y=d*np.random.rand()#position of center of mass of the rod
        theta=np.pi*np.random.rand()#angle of the rod
        if com_y+l*np.sin(theta)/2>d or com_y-l*np.sin(theta)/2<0:#if y_com+l/2*np.sin(theta) exceeds d then the rod crosses the upper strip and the other condition asks whether it crosses lower strip.
            count+=1
    return count/n
        
    
    

    
    
```

```python
d=10
l=5
lens=np.arange(20,200000,5000)#Checking the convergence to get what tis best value for n.
area_list=np.array([prob(i,d,l) for i in lens])
plt.plot(lens,area_list)
plt.xlabel("No of trials")
plt.ylabel("Probability")
```

# We can see that the probability is converging.

```python
N=10**5
n=50
walks=np.zeros(N)
for i in range(n):
    array=np.random.rand(N)#We are simulating a N-random array once to get the position of N random walksers at step n+1 and adding it to walk at step n.
    w=np.where(array<0.5,-1,1)
    walks=walks+w
    
    
    
    
```

```python
def p(x,n):
    return np.exp(-x*x/n)*2/np.sqrt(2*np.pi*n)
    
```

```python
latice=np.arange(-50,51,2.0) #Note that the odd numbered lattices will not be occupied , hence we are removing odd numbered lattices from the X axis.
count=np.zeros(len(latice))
for i in range(len(latice)):
    count[i]=list(walks).count(latice[i])
    
    
    
    
    
```

```python
plt.scatter(latice,count/10**5,color='b',label="Numerical simulation")
plt.plot(latice,p(latice,50),color='r',label="Analytical solution")
plt.xlabel("lattice no")
plt.ylabel("Probability")
```

```python
N=10**5
n=50
walks=np.zeros(N)
x2_avg=np.zeros(n)
for i in range(n):
    array=np.random.rand(N)
    w=np.where(array<0.5,-1,1)
    walks=walks+w
    x2_avg[i]=np.sum(walks*walks)/10**5 #In the previous code ,we just added an extra line to compute x^2_avg at each time n.
    
```

```python
plt.plot(x2_avg,label="<X**2>")
plt.xlabel("No of steps")
plt.ylabel("<x**2>")
plt.legend()
```

# We can see that the dependence of X^2_avg is linear in number of steps.
