```python
maxiter=100
def f(x):
    return (x+3)*(x-1)**2
l,r=-4,-2.5
def swap(l,r):
    if f(l)*f(r)<0:
        c=(l+r)/2
        if f(c)*f(l)<0:
            r=c
        else:
            l=c
        return [l,r]
            
        

def bisect(l,r,i,tol):
    while abs(l-r)>tol and i<maxiter:
            if f(l)==0:
                return [l,i]
            elif f(r)==0:
                return [r,i]
            elif f(l)*f(r)<0:
                return bisect(swap(l,r)[0],swap(l,r)[1],i+1,tol)
            i=i+1
    return [l,r,i+1]
            
def bisection(l,r,tol):
    return bisect(l,r,1,tol)

    

           
            
        
                
        
```

# Secant:

```python
def swap(l,r):
        c=(-r*f(l)+l*f(r))/(f(r)-f(l))
        return c
def secant(l,r,i):
    while abs(r-l)>tol:
        return secant(r,swap(l,r),i+1)
    return r
    
        
        

            
```

```python
print("The four roots are ",secant(-1,-1.5,1),",",secant(-0.1,-0.5,1),",",secant(0.3,0.5,1),",",secant(0.7,0.8,1))
```

# Newton-Raphson Method:

```python
def fp(x):
    return 4*x**3+8*x**2/9-2*x
def update(x):
    return x-f(x)/fp(x)
def newton(l):
    if fp(l)==0:
        return "Function has an extrema"
    while abs(update(l)-l)>tol:
        return newton(update(l))
    return update(l)
```

```python
print("The four roots are ",newton(-1.5),",",newton(-0.5),",",newton(0.3),",",newton(0.7))
```

# Regula Falsi:

```python
def swap(l,r):
    if f(l)*f(r)<0:
        c=(-r*f(l)+l*f(r))/(f(r)-f(l))
        if f(c)*f(l)<0:
            r=c
        else:
            l=c
        return [l,r]
def regulam(l,r,i,tol):
    while abs(l-r)>tol:
            if f(l)==0:
                return [l,i]
            elif f(r)==0:
                return [r,i]
            elif f(l)*f(r)<0:
                return bisect(swap(l,r)[0],swap(l,r)[1],i+1,tol)
            i=i+1
    return [l,r,i+1]
            
def regula(l,r):
    return regulam(l,r,1,10**(-5))[0]

        
```

```python
print("The three roots are ",regula(-1.5,-1),",",regula(-0.5,-0.2),",",regula(0.0,0.5),",",regula(0.7,0.8))
```
